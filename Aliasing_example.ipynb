Example of Aliasing in time domain

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the vibration signal
sampling_rate = 1000  # Sampling rate in Hz
duration = 1.0       # Duration of the signal in seconds
frequencies = [50, 300, 400]  # Frequencies of the vibration components in Hz

# Generate the vibration signal
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
vibration_signal = np.sum([np.sin(2 * np.pi * freq * t) for freq in frequencies], axis=0)

# Plot the original vibration signal
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, vibration_signal)
plt.title("Original Vibration Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Simulate undersampling to demonstrate aliasing
undersampled_rate = 300  # Undersampled rate in Hz
undersampled_indices = np.arange(0, len(t), int(sampling_rate / undersampled_rate))
undersampled_signal = vibration_signal[undersampled_indices]

# Plot the undersampled signal
plt.subplot(2, 1, 2)
plt.plot(t[undersampled_indices], undersampled_signal, 'r')
plt.title("Undersampled Vibration Signal (Aliased)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```

Example of vibration with three modes in time domain and frequency domain

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the vibration signal
sampling_rate = 1000  # Sampling rate in Hz
duration = 1.0       # Duration of the signal in seconds
frequencies = [50, 300, 400]  # Frequencies of the vibration components in Hz

# Generate the vibration signal
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
vibration_signal = np.sum([np.sin(2 * np.pi * freq * t) for freq in frequencies], axis=0)

# Compute the FFT of the vibration signal
fft_freqs = np.fft.fftfreq(len(vibration_signal), d=1/sampling_rate)
fft_values = np.fft.fft(vibration_signal)
fft_magnitude = np.abs(fft_values)

# Plot the original signal in the time domain
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(t, vibration_signal)
plt.title("Original Vibration Signal")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")

# Plot the frequency domain representation
plt.subplot(2, 1, 2)
plt.plot(fft_freqs, fft_magnitude)
plt.title("Frequency Domain Representation")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, sampling_rate / 2)  # Display only positive frequencies
plt.tight_layout()

plt.show()
```

Vibration with noise and example of aliasing

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the vibration signal
sampling_rate = 1000  # Sampling rate in Hz
duration = 1.0       # Duration of the signal in seconds
frequencies = [50, 250, 400]  # Frequencies of the vibration components in Hz

# Generate the vibration signal
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
vibration_signal = np.sum([np.sin(2 * np.pi * freq * t) for freq in frequencies], axis=0)

# Adding Gaussian noise to the vibration signal
noise = np.random.normal(0, 0.5, len(vibration_signal))
noisy_vibration_signal = vibration_signal + noise

# Compute the FFT of the noisy vibration signal
fft_freqs = np.fft.fftfreq(len(noisy_vibration_signal), d=1/sampling_rate)
fft_values = np.fft.fft(noisy_vibration_signal)
fft_magnitude = np.abs(fft_values)

# Simulate undersampling to demonstrate aliasing
undersampled_rate = 300  # Undersampled rate in Hz
undersampled_indices = np.arange(0, len(t), int(sampling_rate / undersampled_rate))
undersampled_signal = noisy_vibration_signal[undersampled_indices]

# Compute the FFT of the undersampled signal
fft_freqs_undersampled = np.fft.fftfreq(len(undersampled_signal), d=1/sampling_rate)
fft_values_undersampled = np.fft.fft(undersampled_signal)
fft_magnitude_undersampled = np.abs(fft_values_undersampled)

# Plot the original and undersampled FFT plots
plt.figure(figsize=(10, 10))

# Original FFT plot
plt.subplot(2, 1, 1)
plt.plot(fft_freqs, fft_magnitude)
plt.title("Frequency Domain Representation (Original + Noise Signal)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, sampling_rate / 2)  # Display only positive frequencies

# Undersampled FFT plot
plt.subplot(2, 1, 2)
plt.plot(fft_freqs_undersampled, fft_magnitude_undersampled, 'r')
plt.title("Frequency Domain Representation (Undersampled + Noise Signal)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, sampling_rate / 2)  # Display only positive frequencies
plt.tight_layout()

plt.show()
```

The example of filter

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, lfilter

# Parameters for the vibration signal
sampling_rate = 1000  # Sampling rate in Hz
duration = 1.0       # Duration of the signal in seconds
frequencies = [50, 250, 400]  # Frequencies of the vibration components in Hz

# Generate the vibration signal
t = np.linspace(0, duration, int(sampling_rate * duration), endpoint=False)
vibration_signal = np.sum([np.sin(2 * np.pi * freq * t) for freq in frequencies], axis=0)

# Adding Gaussian noise to the vibration signal
noise = np.random.normal(0, 0.5, len(vibration_signal))
noisy_vibration_signal = vibration_signal + noise

# Design a low-pass filter
cutoff_frequency = 200  # Cutoff frequency in Hz
b, a = butter(N=6, Wn=cutoff_frequency / (0.5 * sampling_rate), btype='low')

# Apply the low-pass filter to the noisy signal
filtered_signal = lfilter(b, a, noisy_vibration_signal)

# Compute the FFT of the noisy vibration signal
fft_freqs0 = np.fft.fftfreq(len(noisy_vibration_signal), d=1/sampling_rate)
fft_values0 = np.fft.fft(noisy_vibration_signal)
fft_magnitude0 = np.abs(fft_values0)

# Compute the FFT of the filtered signal
fft_freqs = np.fft.fftfreq(len(filtered_signal), d=1/sampling_rate)
fft_values = np.fft.fft(filtered_signal)
fft_magnitude = np.abs(fft_values)

# Simulate undersampling to demonstrate aliasing
undersampled_rate = 300  # Undersampled rate in Hz
undersampled_indices = np.arange(0, len(t), int(sampling_rate / undersampled_rate))
undersampled_signal = filtered_signal[undersampled_indices]

# Compute the FFT of the undersampled signal
fft_freqs_undersampled = np.fft.fftfreq(len(undersampled_signal), d=1/sampling_rate)
fft_values_undersampled = np.fft.fft(undersampled_signal)
fft_magnitude_undersampled = np.abs(fft_values_undersampled)

# Plot the original, filtered, and undersampled FFT plots
plt.figure(figsize=(12, 12))

# Original FFT plot
plt.subplot(3, 1, 1)
plt.plot(fft_freqs0, fft_magnitude0)
plt.title("Frequency Domain Representation (Original + Noise Signal)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, sampling_rate / 2)  # Display only positive frequencies

# Filtered FFT plot
plt.subplot(3, 1, 2)
plt.plot(fft_freqs, np.abs(np.fft.fft(filtered_signal)))
plt.title("Frequency Domain Representation (Filtered Signal)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, sampling_rate / 2)  # Display only positive frequencies

# Undersampled FFT plot
plt.subplot(3, 1, 3)
plt.plot(fft_freqs_undersampled, fft_magnitude_undersampled, 'r')
plt.title("Frequency Domain Representation (Undersampled + Filtered Signal)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.xlim(0, sampling_rate / 2)  # Display only positive frequencies

plt.tight_layout()
plt.show()
```
