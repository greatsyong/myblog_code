```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate a clean signal
t = np.linspace(0, 2*np.pi, 200)
true_signal = np.sin(t)

# 2. Add noise (simulate sensor noise)
np.random.seed(0)
noise = np.random.normal(0, 0.3, size=len(t))
noisy_signal = true_signal + noise

# 3. Moving Average Filter
window_size = 10
kernel = np.ones(window_size) / window_size
filtered_signal = np.convolve(noisy_signal, kernel, mode='same')

# 4. Plot
plt.figure()

plt.plot(t, true_signal, label='True Signal')
plt.plot(t, noisy_signal, label='Noisy Signal')
plt.plot(t, filtered_signal, label='Moving Average (N=10)')

plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.title('Moving Average Filter: Noise Reduction and Smoothing')

plt.legend()
plt.grid()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a clean signal
t = np.linspace(0, 2*np.pi, 200)
true_signal = np.sin(t)

# Add noise
np.random.seed(0)
noise = np.random.normal(0, 0.3, size=len(t))
noisy_signal = true_signal + noise

# -----------------------------
# Figure 1: Basic MAF example
# -----------------------------
window_size = 10
kernel = np.ones(window_size) / window_size
filtered_signal = np.convolve(noisy_signal, kernel, mode='same')

plt.figure(figsize=(10, 5))
plt.plot(t, true_signal, label='True Signal')
plt.plot(t, noisy_signal, label='Noisy Signal')
plt.plot(t, filtered_signal, label='Moving Average (N=10)')
plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.title('Moving Average Filter: Noise Reduction and Smoothing')
plt.legend()
plt.grid()
plt.savefig('maf_basic.png', dpi=300, bbox_inches='tight')
plt.show()

# -----------------------------------
# Figure 2: Window size comparison
# -----------------------------------
plt.figure(figsize=(10, 5))
plt.plot(t, true_signal, label='True Signal')

for w in [3, 10, 20]:
    kernel = np.ones(w) / w
    filtered = np.convolve(noisy_signal, kernel, mode='same')
    plt.plot(t, filtered, label=f'MAF (N={w})')

plt.xlabel('Time')
plt.ylabel('Signal Value')
plt.title('Effect of Window Size on Moving Average Filter')
plt.legend()
plt.grid()
plt.savefig('maf_window_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
